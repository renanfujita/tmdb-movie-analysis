### Análise de Dados Faltantes: Orçamento e Receita

Após o tratamento inicial, observamos que as colunas `Orçamento` e `Receita` ainda contêm um número significativo de valores nulos. Aproximadamente **30% dos filmes** no nosso dataset final não possuem informações financeiras divulgadas.

**Decisão:** Para as análises financeiras (como Receita Líquida), esses filmes serão desconsiderados para não distorcer as médias e os totais. No entanto, eles serão mantidos para outras análises não financeiras, como popularidade por gênero ou avaliação.
